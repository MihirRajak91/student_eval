```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
df['math_score_bins'] = pd.cut(df['math score'], bins=bins, labels=labels, right=False)
sns.countplot(x='math_score_bins', data=df, order=labels)
plt.title('Count of Students by Math Score Range')
plt.xlabel('Math Score Range')
plt.ylabel('Count')
plt.show()
```

```python
df['overall_passstatus'] = df.apply(lambda x : 'F' if x['Math_PassStatus'] == 'F' or x['reading_PassStatus'] == 'F' or x['writing_PassStatus'] =='F' else 'P', axis =1)
```

```python
df['Total_Marks'] = df['math score']+df['reading score']+df['writing score']
df['Percentage'] = df['Total_Marks']/3 
```

```python
df['Percentage']
```

```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
df['Percentage_score_bins'] = pd.cut(df['Percentage'], bins=bins, labels=labels, right=False)
sns.countplot(x='writing_score_bins', data=df, order=labels)
plt.title('Count of Students by Percentage Range')
plt.xlabel('Percentage Range')
plt.ylabel('Count')
plt.show()
```

```python
def GetGrade(Percentage, OverAll_PassStatus):
    if ( OverAll_PassStatus == 'F'):
        return 'F'    
    if ( Percentage >= 80 ):
        return 'A'
    if ( Percentage >= 70):
        return 'B'
    if ( Percentage >= 60):
        return 'C'
    if ( Percentage >= 50):
        return 'D'
    if ( Percentage >= 40):
        return 'E'
    else: 
        return 'F'

df['Grade'] = df.apply(lambda x : GetGrade(x['Percentage'], x['overall_passstatus']), axis=1)

df.Grade.value_counts()
```

```python
sns.countplot(x="Grade", data = df, order=['A','B','C','D','E','F'],  palette="muted")
plt.show()
```

```python
p = sns.countplot(x='parental level of education', data = df, hue='Grade', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
```
